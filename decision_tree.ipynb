# 1. Packages

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, f1_score
from sklearn.tree import DecisionTreeClassifier
from decision_tree import DecisionTree
from util import *

%matplotlib inline
```

# 2. Dataset

I loaded all the datasets to be used.

```python
X1, y1 = load_data('data/processed/dataset_1.csv')
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)

X2, y2 = load_data('data/processed/dataset_2.csv')
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

X3, y3 = load_data('data/processed/dataset_3.csv')
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
```

Then, I checked the dimensions of each datasets.

- First dataset:

```python
print("The first dataset has {} samples and {} features.".format(X1.shape[0], X1.shape[1]))
print("The training set of the first dataset has {} samples.".format(X1_train.shape[0]))
```

- Second dataset:

```python
print("The second dataset has {} samples and {} features.".format(X2.shape[0], X2.shape[1]))
print("The training set of the second dataset has {} samples.".format(X2_train.shape[0]))
```

- Third dataset:

```python
print("The third dataset has {} samples and {} features.".format(X3.shape[0], X3.shape[1]))
print("The training set of the third dataset has {} samples.".format(X3_train.shape[0]))
```

# 3. Models

## 3.1. First dataset

```python
my_dt1 = DecisionTree(min_samples=2, max_depth=5)
my_dt1.fit(X1_train, y1_train)
```

```python
y1_pred = my_dt1.predict(X1_test)
print("The accuracy of the decision tree on the first dataset is {}".format(accuracy_score(y1_test, y1_pred)))
print("The F1 score of the decision tree on the first dataset is {}".format(f1_score(y1_test, y1_pred)))
```

```python
my_cm1 = confusion_matrix(y1_test, y1_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=my_cm1,display_labels=['F','T'])
disp.plot(cmap='Blues')
plt.title('My Decision Tree || Confusion Matrix')
plt.show()
```

```python
dt1 = DecisionTreeClassifier(min_samples_split=2, max_depth=5)
dt1.fit(X1_train, y1_train)
```

```python
y1_real_pred = dt1.predict(X1_test)
print("The accuracy of the sklearn decision tree on the first dataset is {}".format(accuracy_score(y1_test, y1_real_pred)))
print("The F1 score of the sklearn decision tree on the first dataset is {}".format(f1_score(y1_test, y1_real_pred)))
```

```python
cm1 = confusion_matrix(y1_test, y1_real_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=my_cm1,display_labels=['F','T'])
disp.plot(cmap='Blues')
plt.title('My Logistic Regression || Confusion Matrix')
plt.show()
```
